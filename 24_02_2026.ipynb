```python
#descriptive stats
import pandas as pd
data={
    "Ages":[22,25,47,35,46,50,28,33,39,41],
    "Salaries":[3000,3200,5500,4000,5200,6000,3500,3700,4500,4800]
}
df=pd.DataFrame(data)
#apply statistical function
print("summary statistics:\n",df.describe())
#generates summary statistics
print("\n mean:\n",df.mean())
print("\n median:\n",df.median())
print("\n mode:\n",df.mode())
print("\n minimum value:\n",df.min())
print("\n maximum value:\n",df.max())
print("\n standard deviation:\n",df.std())
print("\n variance:\n",df.var())
print("\n count:\n",df.count())
print("\n skewness:\n",df.skew())
print("\n kurtosis:\n",df.kurt())
```

```python
import pandas as pd
#create a sample dataset with missing value
data={
    "name":["alice","bob","charlie","david","eve"],
    "age":[25,None,35,40,28],# bob has missing age
    "salary":[5000,6000,None,7000,6200],#charlie has missing salary
    "allowances":[2000,None,2000,2000,2500] #bob has missing allowances
}
df=pd.DataFrame(data)
#checking for missing values in each column
print("missing values count:\n",df.isnull().sum())
#dropping rows with any missing value
df_cleaned=df.dropna()
print("\n DataFrame after dropping missing values:\n",df_cleaned)
#filling missing values with specific values
df_filled=df.fillna({
    "age":df['age'].mean(),
    "salary":df['salary'].mean(),
    "allowances":2000

})
print("\n DataFrame after filling missing values:\n",df_filled)
```
