```python
#Data manipulation
import pandas as pd
data={
    "Name":['A','B','C'],
    "Age":[20,21,22],
    "Marks":[70,80,90]
}
df=pd.DataFrame(data)
print("original data")
print(df)
df["Bonus"]=5
print("\n after adding Bonus:")
print(df)
df["Marks"]=df["Marks"]+2
print("\n after increasing Marks:")
print(df)
df["Total"]=df["Marks"]+df["Bonus"]
print("\n after adding total:")
print(df)
df.loc[0,"Age"]=21
print("\n After changing age of first std")
print(df)
df.rename(columns={"total":"final"},inplace=True)
print(df)
df.drop("Bonus",axis=1,inplace=True)
print(df)
```

```python

# Selection Sort
n = int(input("Enter no of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))
print("\nOriginal array:", arr)
for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
print("Sorted array:", arr)
```

```python

# Bubble Sort
n = int(input("Enter no of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))
for i in range(n - 1):
    for j in range(n - 1 - i):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
print("Sorted array:", arr)
```

```python
# Binary Search
n = int(input("Enter no. of elements: "))
arr = []
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))
arr.sort()
print("Sorted array:", arr)
key = int(input("Enter element to search: "))
low = 0
high = n - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print(f"Element {key} found at position {mid + 1}")
        found = True
        break
    elif key < arr[mid]:
        high = mid - 1
    else:
        low = mid + 1
if not found:
    print("Element not found")
```

```python

# Grouping
import pandas as pd
# Simple Dataframe
data = {
    "Department": ["IT", "IT", "HR", "HR", "Finance"],
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [80, 85, 75, 90, 88]
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
print("\n Average Marks by department:")
print(df.groupby("Department")["Marks"].mean())
print("\nTotal marks by department:")
print(df.groupby("Department")["Marks"].sum())
```

```python
# Rearrange
import pandas as pd
data = {
    "Name": ["A", "B", "C"],
    "Age": [22, 21, 23],
    "Marks": [85, 90, 80]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
# Rearranging rows by Marks (ascending)
df_sorted = df.sort_values(by="Marks")
print("\nAfter rearranging rows by marks (ascending):")
print(df_sorted)
# Rearranging rows by Marks (descending)
df_sorted_desc = df.sort_values(by="Marks", ascending=False)
print("\nAfter rearranging rows by marks (descending):")
print(df_sorted_desc)
# Rearranging columns
df_rearranged_cols = df[["Marks", "Name", "Age"]]
print("\nAfter rearranging columns:")
print(df_rearranged_cols)
```

```python
# Ranking
import pandas as pd
data = {
    "Name": ["A", "B", "C", "D"],
    "Marks": [85, 92, 78, 80]
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
df["Rank"] = df["Marks"].rank(ascending=False)
print("\nAfter ranking:")
print(df)

```
