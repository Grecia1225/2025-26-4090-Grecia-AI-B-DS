```python
#Download and save the dataset in the local drive
import pandas as pd
#url of dataset(sample csv dataset)
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/iris.csv"
#read dataset from url
df=pd.read_csv(url)
#display first few rows
print(df.head())
#save dataset to local drive
df.to_csv("iris.csv",index=False)
print("dataset saved successfully")
```

```python
#List
superstore=["metro",'tesco','walmart','kmart','carrefour']
print(superstore)
number_list=[4,4,6,7,2,9,10,15]
mixed_list=['cat',87.23,65,[9,7,3,3]]
print(type(mixed_list))
#list operations
list1=[3,5,7,1]
list2=[2,4,6,8]
print(list1+list2) #list concartination
print(list1*3)
print(list1==list2)
#string to list
quote='Hello World'
string_to_list=list(quote)
print(string_to_list)
# list indexing, slicing
print(list1[:2])
print(list1[-2:])
print(5 in list1)
print(11 in list1)
#list built-in functions
cars=['BMW','Audi','Mercedes','Toyota']
print(len(cars))
print(max(number_list))
print(min(number_list))
print(sum(number_list))
print(sorted(list1))
print(sorted(list1,reverse=True))
#list methods
cars.append('Tesla')
print(cars)
cars.insert(2,'Ford')
print(cars)
cars.remove('Tesla')
print(cars)
cars.pop()
print(cars)
#Nested list
nested_list=[[1,2,3],[4,5,6],[7,8,9]]
print(nested_list)
print(nested_list[1])
print(nested_list[1][2])
```

```python
#Tuples
a=(1,2,3,4,5)
b=(6,7,8,9,10)
print(a+b)
print(a*3)
print(1 in a)
print(11 in a)
print(a[1:4])
print(len(a))
print(max(a))
print(min(a))
print(sum(a))
print(sorted(a))
print(sorted(a,reverse=True))
```

```python
#Sets
a={1,2,3,4,5}
b={3,4,5,6,7}
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.symmetric_difference(b))
print(a.issubset(b))
print(a.isdisjoint(b))
```
