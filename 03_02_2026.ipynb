```python
#sets
A = {10, 4, 8, 1, 5}
B = {4, 5, 6, 7}
A.add(15)              # Adds 15 to set A
A.remove(10)           # Removes 10; raises error if 10 is missing
A.discard(99)          # Safely removes 99; no error if missing
popped = B.pop()       # Removes and returns a random element
union = A | B
inter = A & B
diff = A - B
sym_diff = A ^ B
print(f"Union (|): {union}")
print(f"Intersection (&): {inter}")
print(f"Difference (-): {diff}")
print(f"Symmetric Difference (^): {sym_diff}")
```
