```python
#selection of data

#import pandas and create simple data
#selection particular column and rows
import pandas as pd
data={
    'ID':[101,102,103,104,105],
    'Name':['A','B','C','D','E'],
    'Age':[20,21,19,17,18],
    'marks':[85,78,92,88,90]
}
df=pd.DataFrame(data)
print(df)
print(df.loc[1:3,('Name','Age')])
print(df['marks'])
print(df.iloc[0:3, 1:4])
print(df.loc[2, 'marks'])
print(df[df['Age'] > 18])
print(df[['Name', 'marks']])
df['Passed'] = df['marks'] >= 80
print(df)
print(df.sort_values(by='marks', ascending=False))
print(df['marks'].mean())
```

```python
# filter data
filtered_age = df[df['Age'] > 18]
print(filtered_age)
filtered_marks = df[df['marks'] > 85]
print(filtered_marks)
filtered_multiple = df[(df['Age'] > 18) & (df['marks'] > 85)]
print(filtered_multiple)
filtered_or = df[(df['Age'] > 18) | (df['marks'] > 90)]
print(filtered_or)
filtered_loc = df.loc[df['marks'] >= 80, ['Name', 'marks']]
print(filtered_loc)
filtered_isin = df[df['Name'].isin(['A', 'C', 'E'])]
print(filtered_isin)
```

```python
#data manipulation
import pandas as pd
data = {
    'ID': [101, 102, 103, 104, 105],
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [20, 21, 19, 17, 18],
    'Marks': [85, 78, 92, 88, 90]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
print("\nStudents with Age > 18:")
print(df[df['Age'] > 18])
print("\nStudents with Marks >= 85:")
print(df[df['Marks'] >= 85])
print("\nName and Marks:")
print(df[['Name', 'Marks']])
df.loc[df['Marks'] < 80, 'Marks'] = 80
print("\nUpdated Marks (minimum 80):")
print(df)
df['Result'] = df['Marks'] >= 80
print("\nAdded Result column:")
print(df)
print("\nSorted by Marks (Descending):")
print(df.sort_values(by='Marks', ascending=False))
print("\nAverage Marks:")
print(df['Marks'].mean())
print("\nMaximum Marks:")
print(df['Marks'].max())
```

```python
import pandas as pd
data = {
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Age': [20, 21, 19, 22, 18],
    'Marks': [85, 78, 92, 88, 90]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
group_mean = df.groupby('Department')['Marks'].mean()
print("\nAverage Marks by Department:")
print(group_mean)
group_stats = df.groupby('Department')['Marks'].agg(['mean', 'max', 'min'])
print("\nMarks Statistics by Department:")
print(group_stats)
group_count = df.groupby('Department')['Name'].count()
print("\nStudent Count by Department:")
print(group_count)
group_multi = df.groupby(['Department', 'Age'])['Marks'].mean()
print("\nGrouped by Department and Age:")
print(group_multi)
```

```python
# Selection Sort Program
arr = [64, 25, 12, 22, 11]
n = len(arr)
for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]
print("Sorted array:")
print(arr)
```

```python
# Bubble Sort Program
arr = [64, 34, 25, 12, 22, 11, 90]
n = len(arr)
for i in range(n - 1):
    for j in range(n - i - 1):
        if arr[j] > arr[j + 1]:
            # swap
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
print("Sorted array:")
print(arr)
```

```python
# Linear Search Program
arr = [10, 25, 30, 45, 50]
key = 30
found = False
for i in range(len(arr)):
    if arr[i] == key:
        print("Element found at index:", i)
        found = True
        break
if not found:
    print("Element not found")
```

```python
# Binary Search Program
arr = [10, 20, 30, 40, 50, 60]
key = 40
low = 0
high = len(arr) - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print("Element found at index:", mid)
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1
if not found:
    print("Element not found")
```
