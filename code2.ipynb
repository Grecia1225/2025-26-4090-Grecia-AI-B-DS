```python
import numpy as np

# Create a 1D NumPy array representing daily sales figures
daily_sales = np.array([120, 150, 90, 200, 175, 130, 160])

# Calculate basic statistics
min_sales = daily_sales.min()
max_sales = daily_sales.max()
mean_sales = daily_sales.mean()
total_days = daily_sales.size

# Print the results
print("Minimum sales:", min_sales)
print("Maximum sales:", max_sales)
print("Average sales:", mean_sales)
print("Total number of days:", total_days)
```

```python
import numpy as np

# Create a 2D NumPy array representing temperature readings
temperature_grid = np.array([
    [30, 32, 31, 29],
    [28, 27, 26, 25],
    [33, 34, 35, 36],
    [22, 23, 24, 21]
])

# Slice: extract the top-left 2x2 subgrid
top_left = temperature_grid[0:2, 0:2]

# Reshape: convert the 2x2 subgrid into a 1D array
reshaped = top_left.reshape(-1)

# Print results
print("Original Grid:\n", temperature_grid)
print("Sliced Subgrid:\n", top_left)
print("Reshaped Array:", reshaped)
```

```python
import pandas as pd

# Sample product inventory data
inventory_data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer'],
    'Stock': [5, 25, 8, 12, 3],
    'Price': [1000, 25, 45, 150, 120]
}

# Create the DataFrame
df = pd.DataFrame(inventory_data)

# Filter: products with stock less than 10 AND price greater than 50
low_stock_high_price = df[(df['Stock'] < 10) & (df['Price'] > 50)]

# Display the filtered results
print("Filtered Inventory:\n", low_stock_high_price)
```
